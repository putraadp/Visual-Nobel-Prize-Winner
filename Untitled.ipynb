# Nobel Prize

The Nobel Prize is perhaps the world's most well known scientific award in the world.

Every year it's given to scientists and scholars in the categories chemistry, literature, physics, physiology or medicine, economics, and peace. The first Nobel Prize was handed out in 1901, and at that time the Prize was very Eurocentric and male-focused, but nowadays it's not biased in any way whatsoever.

Surely. Right?
Maybe we can check it out.

### Importing Package and Dataset

The Nobel Foundation has made a dataset available of all prize winners from the start of the prize, in 1901, to 2016. Let's load it in and take a look.

```python
# Import Package
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_style('whitegrid')

# Import Dataset
nobel = pd.read_csv('datasets/nobel.csv')
nobel.head()
```

We can see here, The first five nobel prize is a Man from Europe with different country. But that was back in 1901, looking at all winners in the dataset, from 1901 to 2016, which sex and which country is the most commonly represented?

We can see too, there are so many missing values in dataset. But, at this time we cant fill the missing values with math or estimate because Nobel Prize is open for everyone, even if the person is not from a formally educated circle, and its not a problem for our project.

### Sex and Country is The Most Commonly Represented

First, let's calculate how many Nobel Winners from 1901 to 2016?

```python
print('Number of Nobel Winners :', len(nobel))
```

There are 911 Nobel Winners but, 1 prize can be take by more than 1 people. So, lets count how many Male and Female (sex) winning Nobel Prize.

```python
bysex_winner = nobel.sex.value_counts()

print(bysex_winner)
print('Total :',bysex_winner.sum())
```

Here we know, there are 26 Nobel Prize is sharring or taked by more than 1 people.

Lets find out, which country is the most Nobel Prizer.

```python
country_winner = nobel.birth_country.value_counts()

print(country_winner)
print('Total :', country_winner.sum())
```

From the dataset we know a man and USA is the common winner of Nobel Prize. But, does a man from USA is dominating Nobel Prize?

```python
usa_man = nobel[(nobel.sex == 'Male') & (nobel.birth_country == 'United States of America')]

print(len(usa_man))
```

From 259 prize winners from USA, 248 winners is a man. So, yes a man from USA is dominating Nobel Prize winner from 1901 to 2016.

### USA Starting The Dominance 

Not so surprising perhaps, the most common Nobel laureate between 1901 and 2016 was a man born in the United States of America. But in 1901 all the winners were European.

When did the USA start to dominate the Nobel Prize charts?

```python
nobel['usa_born_winner'] = nobel['birth_country'] == "United States of America"
nobel['decade'] = (np.floor(nobel['year']/10)*10).astype(int)

#proportion of USA winners
prop_uw = nobel.groupby(['decade'], as_index=False)['usa_born_winner'].mean()
prop_uw
```

To make it easier to see, lets find out with plot.

```python
ax = sns.lineplot(data=prop_uw,x='decade',y='usa_born_winner')

from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter(1.0))
plt.ylabel('USA - Nobel Winners')
plt.xlabel('Decade')
plt.show()
```

From graph we can know, the dominance of USA is started from decade 1930s, until 2016.

### What is the gender of a typical Nobel Prize winner?
So the USA became the dominating winner of the Nobel Prize first in the 1930s and had kept the leading position ever since. But one group that was in the lead from the start, and never seems to let go, are a <em>men</em>. Maybe it shouldn't come as a shock that there is some imbalance between how many male and female prize winners there are, but how significant is this imbalance? And is it better or worse within specific prize categories like physics, medicine, literature, etc.?

```python
nobel['female_winner'] = nobel['sex']=='Female'
prop_female_winners = nobel.groupby(['decade','category'], as_index = False)['female_winner'].mean()

ax = sns.lineplot(x='decade',y='female_winner',data=prop_female_winners,hue='category')
ax.yaxis.set_major_formatter(PercentFormatter(1.0))
plt.ylabel('Female Winners')
plt.xlabel('Decade')
plt.show()
```

in totals, womens commonly be a Nobel Prize winner on Peace and Literature category.

### Repeating Winners

For most scientists/writers/activists a Nobel Prize would be the crowning achievement of a long career. But for some people, maybe one is just not enough, and few have gotten it more than once.

So, who are these lucky few?

```python
morethan = nobel.groupby('full_name').filter(lambda group:len(group) >= 2)
morethan
```

```python
print('People with more than 1 Nobel Prize :',len(morethan.full_name.unique()))

morethan.set_index(['full_name','category']).sort_values(by='full_name')
```

There a several people, actually 6 people who win twice and more times Nobel Prize.

We can see to Marie Curie, the first women Nobel winner, show up with 2 nobel price from different category, Physics in 1903 and Chemistry in 1911.
